```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
events = pd.DataFrame({
    "event_id": [1,2,3,4,5],
    "title": [
        "AI Workshop",
        "Machine Learning Seminar",
        "Football Tournament",
        "Music Festival",
        "Data Science Bootcamp"
    ],
    "description": [
        "artificial intelligence machine learning data",
        "deep learning neural network ai",
        "football sport competition team",
        "music concert entertainment singer",
        "data science python machine learning"
    ]
})

events
```

```python
vectorizer = TfidfVectorizer()
```

```python
tfidf_matrix = vectorizer.fit_transform(events["description"])
```

```python
tfidf_matrix.toarray()
```

```python
liked_events = [0,4]   # index của event user thích

liked_events
```

```python
student_profile = tfidf_matrix[liked_events].mean(axis=0)

student_profile = np.asarray(student_profile)
```

```python
similarities = cosine_similarity(student_profile, tfidf_matrix)

similarities
```

```python
scores = list(enumerate(similarities.flatten()))

scores
```

```python
sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)

sorted_scores
```

```python
recommended = []

for index, score in sorted_scores:
    
    if index not in liked_events:
        recommended.append(events.iloc[index]["title"])
    
    if len(recommended) == 3:
        break

recommended
```

```python
print("Recommended events for student:\n")

for event in recommended:
    print("-", event)
```
