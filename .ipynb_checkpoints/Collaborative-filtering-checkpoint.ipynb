```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings = pd.DataFrame({
    "user": ["A","A","B","B","C","C"],
    "event": ["AI","Football","AI","Music","Football","Music"],
    "rating": [5,3,4,2,5,4]
})

ratings
```

```python
matrix = ratings.pivot_table(
    index="user",
    columns="event",
    values="rating"
).fillna(0)

matrix
```

```python
user_similarity = cosine_similarity(matrix)

user_similarity
```

```python
similarity_df = pd.DataFrame(
    user_similarity,
    index=matrix.index,
    columns=matrix.index
)

similarity_df
```

```python
similar_users = similarity_df["A"].sort_values(ascending=False)

similar_users
```

```python
most_similar_user = similar_users.index[1]

most_similar_user
```

```python
matrix.loc[most_similar_user]
```

```python
recommendation = matrix.loc[most_similar_user]

recommendation
```
